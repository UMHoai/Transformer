# Feature Selection

Select important features from entire dataset. There are lot of techniques to select features. This technique helps alot like:
- Reduce Dataset size
- Reduce trainning time
- Enhance Training process
- Accelerate training acuracy

**Methods of feature selection**
- using Machine learning model / embeded
- filter method
- wrapper
- 

**Filter method** : using variance , Threshold, Correlation.

**Wrapper method** : Forward selection, Backward selection, Recursive feature selection

**Embeded** : using any Machine learning model and train the model , finally get feature_importance function


[Broadly Discussed in](https://machinelearningmastery.com/feature-selection-with-real-and-categorical-data/)

*I'm using Embeded method*

```python
importances = feature.feature_importances_

indices = np.argsort(importances)[::-1]

print('Feature Ranking:')

for f in range(len(data.columns[:-1])):
    print('%d. features %d (%f)'% (f+1, indices[f], importances[indices[f]]))
```

```python
indices[:11]
```

```python
features = data.iloc[:,indices[:11]]
```

```python
from sklearn.model_selection import train_test_split

xtrain,ytrain,xtest,ytest = train_test_split(features,label,test_size=.2)
accuracy =[]
```

```python

from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,xtest)
pred = dt.predict(ytrain)
# print(confusion_matrix(ytest,pred))
acc = accuracy_score(ytest,pred)
accuracy.append(['dt',acc])
report = classification_report(ytest,pred)
print(acc,report)
sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```

```python
rf = RandomForestClassifier()
rf.fit(xtrain,xtest)
pred = rf.predict(ytrain)
# print(confusion_matrix(ytest,pred))
acc = accuracy_score(ytest,pred)
accuracy.append(['rf',acc])
report = classification_report(ytest,pred)

print(acc,report)
sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```

```python
lgb = LGBMClassifier()
lgb.fit(xtrain,xtest)
pred = lgb.predict(ytrain)
# print(confusion_matrix(ytest,pred))

acc = accuracy_score(ytest,pred)
accuracy.append(['lgb',acc])
report = classification_report(ytest,pred)
print(acc,report)

sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```

```python
cat = CatBoostClassifier(iterations=2000)
cat.fit(xtrain,xtest,silent=True)
pred = cat.predict(ytrain)
# print(confusion_matrix(ytest,pred))

acc = accuracy_score(ytest,pred)
accuracy.append(['cat',acc])
report = classification_report(ytest,pred)
print(acc,report)

sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```

```python
xgb = XGBClassifier()
xgb.fit(xtrain,xtest)
pred = xgb.predict(ytrain)
# print(confusion_matrix(ytest,pred))

acc = accuracy_score(ytest,pred)
accuracy.append(['xgb',acc])
report = classification_report(ytest,pred)
print(acc,report)

sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```

```python
accuracy = pd.DataFrame(accuracy,columns = ['name','accuracy'])

accuracy = accuracy.sort_values(by = 'accuracy')

sns.barplot(accuracy['name'],accuracy['accuracy'])
```

```python
from sklearn.model_selection import GridSearchCV


```

```python
params = {'depth'         : [4,5,9, 10],
                 'learning_rate' : [0.01,.1,.001],
                 }
cv = GridSearchCV(CatBoostClassifier(task_type = 'GPU',iterations=1000), param_grid = params, cv=6)
cv.fit(xtrain,xtest,silent = True)
print(cv.best_params_,cv.best_score_)
```

```python
pred = cv.predict(ytrain)
acc = accuracy_score(ytest,pred)
accuracy.append(['cat',acc])
report = classification_report(ytest,pred)
print(acc,report)

sns.heatmap(confusion_matrix(ytest,pred),annot=True,fmt = '.1f')
```
